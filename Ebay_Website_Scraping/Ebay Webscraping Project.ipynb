```python
import numpy as np
import pandas as pd
import bs4
from bs4 import BeautifulSoup as bs
import requests
import warnings
warnings.filterwarnings("ignore")
```

```python
url=r"https://www.ebay.com/e/row/pre-loved-row"
```

```python
r=requests.get(url)
```

```python
htmlcontent=r.content
```

```python
soup=bs(htmlcontent,"lxml")
soup
```

```python
soup.find_all("li",class_="s-item s-item--large")
```

```python
my_list=[]
all_li=soup.find_all("li",class_="s-item s-item--large")
for item in all_li:
    prod_name=item.find("h3",class_="s-item__title").text
    prod_link=item.find("a",class_="s-item__link")["href"]
    prod_price=item.find("span",class_="s-item__price").text
    shipping_price=item.find("span",class_="s-item__shipping s-item__logisticsCost").text
    
    di={
        "Product name ": prod_name,
        "Product Price":prod_price,
        "Product Link":prod_link,
        "Shipping Price":shipping_price
    }
    
    my_list.append(di)
    
print(my_list)
```

```python
link_list=["https://www.ebay.com/e/row/pre-loved-row","https://www.ebay.com/e/row/pre-loved-row?_pgn=2"]
```

```python
product_data=[]
def all_prods():
    for links in link_list:
        r=requests.get(links)
        content=r.content
        soup=bs(content,"lxml")
        
        all_li=soup.find_all("li",class_="s-item s-item--large")
        for item in all_li:
            prod_name=item.find("h3",class_="s-item__title").text
            prod_link=item.find("a",class_="s-item__link")["href"]
            prod_price=item.find("span",class_="s-item__price").text
            shipping_price=item.find("span",class_="s-item__shipping s-item__logisticsCost").text
    
            di={
                "Product name ": prod_name,
                "Product Price":prod_price,
                "Product Link":prod_link,
                "Shipping Price":shipping_price
            }
        
            product_data.append(di)
        
all_prods()        
```

```python
product_data
```

```python
final_data_of_purses=pd.DataFrame(product_data)
```
